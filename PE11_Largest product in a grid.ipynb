```python
import pandas as pd
data = pd.read_csv('numgrid.csv', header = None)
maximum = 1
for i in range(20):
    for j in range(20):
        #left direction
        if j <=16:
            prodcol = data[i][j]*data[i][j+1]*data[i][j+2]*data[i][j+3]
            if prodcol > maximum:
                maximum = prodcol
        #right direction        
        if i <= 16:
            prodrow = data[i][j]*data[i+1][j]*data[i+2][j]*data[i+3][j]
            if prodrow > maximum:
                maximum = prodrow
        #right diagonal     
        if j<=16 and i<=16:
            prodRdiag = data[i][j]* data[i+1][j+1]*data[i+2][j+2]*data[i+3][j+3]
            if prodRdiag > maximum:
                maximum = prodRdiag
        #left diagonal
        if i >=3 and j <= 16:
            prodLdiag = data[i][j]* data[i-1][j+1] * data[i-2][j+2]* data[i-3][j+3]
            if prodLdiag > maximum:
                maximum = prodLdiag                
print(maximum)
```
