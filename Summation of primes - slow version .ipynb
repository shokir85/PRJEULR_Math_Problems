```python
import time

t0 =time.time()

lst = []

summa = 2

for s in range(1, 2000001,2):

    lst.append(s)

    primelist = []

    while s%2 ==0:

        primelist.append(2)

        s/=2

    f = 3

    while s >1:

        if s%f == 0:

            primelist.append(f)

            s /=f

        else:

            f += 2

    if primelist == lst:

        summa = summa + primelist[0]

        print(primelist)

        print(lst, summa)

    lst = []

t1 = time.time()

print(t1-t0)
```
