```python
#PROJECT EULER  PROBLEM 18
import time as t
start = t.time()
import pandas as pd
import numpy as np
data = pd.read_fwf('maxpath.txt', header = None)
for i in range(13,-1,-1):
    for j in range(14,-1,-1):
        if not np.isnan(data[j][i]):
            data[j][i] += max(data[j+1][i+1],data[j][i+1])
print(data[j][i])
print(t.time()-start)
```
