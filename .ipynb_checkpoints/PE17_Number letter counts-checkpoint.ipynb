```python
import time as t
numbers={"1":"one", "2":"two", "3":"three", "4":"four", "5":"five", "6":"six", "7":"seven", "8":"eight", "9":"nine",
             "11" :"eleven", "12" : "twelve", "13":"thirteen","14":"fourteen","15":"fifteen", "16":"sixteen",
             "17":"seventeen", "18":"eighteen", "19":"nineteen","10":"ten", "20":"twenty", "30":"thirty", "40":"forty",
             "50":"fifty", "60":"sixty","70":"seventy","80":"eighty", "90":"ninety","100" : "hundred", 
             '1000':'thousand', 'and':'and'}
t0 = t.time()
count = 0
for m in range(1,1001):    
    n = str(m)
    l = len(n)
    if n in numbers: # checking numbers already existing in dictionary
        if m ==100:
            item = numbers['1']+numbers['100']
            count +=len(item)
        elif m ==1000:
            item = numbers['1']+numbers['1000']
            count +=len(item)
        else:
            item = numbers[n]
            count += len(numbers[n])
    
    elif m < 100:#checking numbers less than 100
        for i in range(l):
            if i == 0:
                item = n[i]+'0' 
                count +=len(numbers[item])
            else:
                item = n[i]
                count += len(numbers[item])
    else:# checking numbers with 3 digits, between 101 and 999 inclusive
        if int(n[1])==0 and int(n[2])==0: # checking numbers with 2 zero digits, such as 200, 300,..,900
            item = numbers[n[0]]+numbers['100']
            count +=len(item)
        else: # checking all remaining numbers
            for i in range(l): 
                num = n[i]
                if i == 0:
                    item = numbers[num]+numbers['100']
                    count +=len(item)
                elif i == 1:
                    if int(num) > 1:
                        item = numbers['and']+numbers[num+'0']
                        count +=len(item)
                    elif int(num) == 0:
                        item = numbers['and'] + numbers[n[i+1]]
                        count +=len(item)
                    elif int(num) == 1:
                        item = numbers['and'] + numbers[n[i]+n[i+1]]
                        count +=len(item)
                else:
                    if int(n[i-1])!= 0 and int(n[i-1])!=1 and int(n[i])!=0:
                        item = numbers[num]
                        count +=len(item)
                    else:
                        continue
print(count)
print(t.time()-t0)
```
