```python
import math
lprob = 0
uprob = 1
one = [1]
zero = [0]
binary = [[1],[0]]
for lst in binary:
    max_sum = (math.pi)**2/6
    current_sum = 0
    dcount = 0
    l = len(lst)
    cprob = 1/(2**l)
    for d in lst:
        dcount += 1
        current_sum += d/dcount**2
        if d == 0:
            c_sum = 1/dcount**2
            max_sum = max_sum - c_sum     
    if current_sum > 0.5:
        lprob += cprob
        #print('lprob=', lprob)
        binary.remove(lst)     
    if max_sum < 0.5:
        uprob = uprob - cprob
        binary.remove(lst)
print('lprob=', lprob) 
print('uprob=', uprob)
```

```python
import math
print('-----------------------------------------------------------------------------------')
lprob = 0
uprob = 1
one = [1]
zero = [0]
binary = [[1],[0]]
lb = len(binary)
count = 0
init_lprob = lprob
init_uprob = uprob
while lb > 1:
    print('-----------------------------------------------------------------------------------')
    count += 1
    print('count:', count)
    print('init_lowerprob:', init_lprob)
    print('init_upperprob:', init_uprob)
    print('starting_binary:', binary)
    print('len_starting_binary:',len(binary))
    
    for lst in binary:
        print('lst:',lst)
        max_sum = (math.pi)**2/6
        current_sum = 0
        dcount = 0
        l = len(lst)
        cprob = 1/(2**l)
        for d in lst:
            dcount += 1
            current_sum += d/dcount**2
            if d == 0:
                c_sum = 1/dcount**2
                max_sum = max_sum - c_sum
        delnumber = 0
        print('current_sum:', current_sum)
        if current_sum > 0.5:
            delnumber += 1
            lprob += cprob
            print('lprob:', lprob)
            print('CURRENT_SUM:', current_sum)
            print('-----remove the list from binary----')
            print('removed_binary_list:', lst)
            binary.remove(lst)     
        print('max_sum:', max_sum)
        if max_sum < 0.5:
            uprob = uprob - cprob
            delnumber += 1
            print('MAX_SUM:', max_sum)
            print('----remove the list from binary----')
            print('removed_binary_list:', lst)
            print('uprob:', uprob)
            binary.remove(lst)
        print('number_removed_lst:', delnumber)
    print('remaining_binary:', binary)
    print('len_remaining_binary:',len(binary))
    lst3 = []
    for lst in binary:
        lst1 = lst + one
        lst2 = lst + zero
        lst3.append(lst1)
        lst3.append(lst2)
    binary = lst3
    init_lprob = lprob
    init_uprob = uprob
    print('new_binary:', binary)
    print('len_new_binary:', len(binary))
    print('count:',count)
    print('final_lprob:', lprob) 
    print('final_uprob:', uprob)
    print('deltaprob:', uprob - lprob)
    print('-----------------------------------------------------------------------------------')
    if count == 1:
        break
```

```python
def add_digits(binary):
    new_list = []
    for lst in binary:
        lst1 = lst + [1]
        lst2 = lst + [0]
        new_list.append(lst1)
        new_list.append(lst2)
    return new_list
```
